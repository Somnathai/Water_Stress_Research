```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('dataset.csv')

# Convert 'TimePeriod' to string
df['TimePeriod'] = df['TimePeriod'].astype(str)

# Filter data for the specified time periods (2000-2010)
df_2000_2010 = df[df['TimePeriod'].between('2000', '2010')]

# Create a pivot table with 'GeoAreaName' as index and 'TimePeriod' as columns
pivoted_df = df_2000_2010.pivot_table(index='GeoAreaName', columns='TimePeriod', values='Value', aggfunc='first')

# Plot bar charts for each 'GeoAreaName'
for geo_area in pivoted_df.index:
    values = pivoted_df.loc[geo_area].tolist()
    years = pivoted_df.columns.tolist()

    # Plot bar chart
    plt.bar(years, values)
    plt.title(f'Values for {geo_area}')
    plt.xlabel('Time Period')
    plt.ylabel('Value')
    plt.show()

    # Calculate the change between the first and last values in the 2000-2010 range
    first_value = values[0]
    last_value = values[-1]
    change = last_value - first_value

    # Determine the message based on the change
    if change == 0:
        message = "Same"
    elif change > 0:
        message = "Increased"
    elif change < 0:
        message = "Decreased"

    print(f"For {geo_area} within the 2000-2010 period: {message}\n")
```
